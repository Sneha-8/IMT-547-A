```python
"""
I4 Easy to Access: 3 different information structures with different access technologies
Author: Sneha

Requirements (already available in Colab):
    pip install requests beautifulsoup4

Run all cells with Runtime > Run all
"""

import requests
from bs4 import BeautifulSoup
import csv
import io


# =============================================================================
# FUNCTION 1: JSON via REST API (HTTP connection to public API)
# =============================================================================
#
# Access Technology: REST API over HTTP
# Information Structure: JSON
#
# Data Source: Open-Meteo (https://open-meteo.com/) - free, no API key needed
# Fetches current weather data for Seattle, WA.
#
# PROS:
#   - Structured, machine-readable data (JSON is easy to parse in Python)
#   - Real-time / up-to-date data straight from the source
#   - No authentication required for many public APIs
#   - Widely supported; requests library makes it simple
#
# CONS:
#   - Requires internet connection at runtime
#   - API may change, go down, or enforce rate limits
#   - Data is only as reliable/complete as the API provider makes it
#   - Some APIs require API keys, registration, or payment
# =============================================================================

def fetch_json_from_api():
    print("=" * 60)
    print("FUNCTION 1: JSON via REST API")
    print("=" * 60)

    url = (
        "https://api.open-meteo.com/v1/forecast"
        "?latitude=47.6062&longitude=-122.3321"
        "&current_weather=true"
    )

    response = requests.get(url)
    response.raise_for_status()

    data = response.json()

    print(f"Location: Seattle, WA (lat={data['latitude']}, lon={data['longitude']})")
    weather = data["current_weather"]
    print(f"Current Temperature: {weather['temperature']} C")
    print(f"Wind Speed:          {weather['windspeed']} km/h")
    print(f"Weather Code:        {weather['weathercode']}")
    print(f"Time:                {weather['time']}")
    print()


# =============================================================================
# FUNCTION 2: CSV via direct file download over HTTP
# =============================================================================
#
# Access Technology: HTTP file download (no special library beyond requests)
# Information Structure: CSV (comma-separated values)
#
# Data Source: GitHub-hosted public dataset - US state populations
# URL: https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv
#
# PROS:
#   - CSV is universally readable (Excel, Google Sheets, Python, R, etc.)
#   - Simple flat structure - easy to parse row by row
#   - Widely available for open/public datasets (e.g. Kaggle, data.gov)
#   - Can be downloaded once and used offline
#
# CONS:
#   - No built-in data types (everything is a string until you convert it)
#   - No schema enforcement - columns can be missing or misnamed
#   - Large CSV files can be slow to read into memory
#   - Not ideal for hierarchical or nested data
# =============================================================================

def fetch_csv_from_url():
    print("=" * 60)
    print("FUNCTION 2: CSV via HTTP File Download")
    print("=" * 60)

    url = (
        "https://raw.githubusercontent.com/jakevdp/"
        "data-USstates/master/state-population.csv"
    )

    response = requests.get(url)
    response.raise_for_status()

    content = response.content.decode("utf-8")
    reader = csv.DictReader(io.StringIO(content))

    rows = list(reader)
    print(f"Total rows in dataset: {len(rows)}")
    print(f"Columns: {', '.join(rows[0].keys())}")
    print("\nFirst 5 rows:")
    for row in rows[:5]:
        print(dict(row))
    print()


# =============================================================================
# FUNCTION 3: HTML via web scraping (requests + BeautifulSoup)
# =============================================================================
#
# Access Technology: HTTP request + HTML parsing with BeautifulSoup
# Information Structure: HTML
#
# Data Source: Wikipedia - List of U.S. states and territories
# URL: https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States
#
# PROS:
#   - Works on any publicly accessible webpage - huge amount of data available
#   - BeautifulSoup makes HTML parsing straightforward
#   - No API or special access required
#   - Can extract structured data from unstructured HTML tables
#
# CONS:
#   - HTML structure can change, breaking your scraper without warning
#   - Some websites block scrapers (rate limiting, CAPTCHAs, JS rendering)
#   - Data may need significant cleaning - HTML is for display, not data exchange
#   - Not all sites allow scraping (check robots.txt and terms of service)
# =============================================================================

def fetch_html_and_scrape():
    print("=" * 60)
    print("FUNCTION 3: HTML via Web Scraping (BeautifulSoup)")
    print("=" * 60)

    url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"

    headers = {"User-Agent": "Mozilla/5.0 (educational assignment)"}
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    print(f"Page title: {soup.title.string}")

    table = soup.find("table", {"class": "wikitable"})
    rows = table.find_all("tr")

    print(f"\nFound table with {len(rows)} rows. First 5 data rows:\n")
    for row in rows[1:6]:
        cells = [cell.get_text(strip=True) for cell in row.find_all(["td", "th"])]
        print(cells)
    print()


# =============================================================================
# MAIN - Run all three functions
# =============================================================================

fetch_json_from_api()
fetch_csv_from_url()
fetch_html_and_scrape()
```
